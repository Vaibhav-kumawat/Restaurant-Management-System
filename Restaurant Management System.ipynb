```python
import json
from datetime import datetime
from time import sleep
from pprint import pprint
```

```python
def staf(data):
    """
        Take details as a input and store in a dictionary.
        First line contain an integer N,to represent number of candidate.
        Input formate like as -->
                                1. Id
                                2. Name of staf
                                3. Post/Possition
                                4. Phone number 
                                5. City
    """
    
    # start your code yera-->
    data = data
    n = int(input("Enter no staf -- > "))
    for i in range(n):
        Id= int(input("Enter your Id --> "))
        name = input("Enter your name --> ").strip().upper()
        Post = input("Enter Post -->").strip().upper()
        phone = int(input("Enter Phone number --> "))
        city = input("Enter City name --> ").strip().upper()

        data[Id] = {}
        data[Id]["name"] = name
        data[Id]["Post"] = Post
        data[Id]["Mob.No"] = phone
        data[Id]["City"] = city

        
    return data
```

```python
menu_card = {
             "Snacks" : {
                            0 : ["Snacks_name","Price","Time_in _making_in_minutes"],
                            1 : ["Pizza",120,20],
                            2 : ["Burger",50,15],
                            3 : ["Veg_roll",60,10],
                            4 : ["Pasta",80,15],
                            5 : ["Chow_mein",45,10]
                        },
             "Drinks" : {
                            0 : ["Drinks_name","Price","Time_in _making_in_minutes"],
                            1 : ["Cold_coffee",80,10],
                            2 : ["Hot_coffee",45,10],
                            3 : ["Coka_cola",60,5]
                         },
             "Desserts" : {
                             0 : ["Desserts_name","Price","Time_in _making_in_minutes"],
                             1 : ["Balu shahi",120,30],
                             2 : ["Gajar ka Halwa",200,30],
                             3 : ["Gulab jamun",250,30]
                          }
            }
```

```python
def Customer(menu_card : dict, Customers_record : dict):
    print("-"*80)
    print("Welcom in Family Restaurant".center(80))
    print("-"*80)
    sleep(2)
    for i in range(10):
        print("\nOptions --> \n1. Order_food \n2. Check_menu \n3. Show_my_cart \n4. Exit")
        sleep(0.5)
        n = int(input("Choose any option from above --> "))
        orders = []
        sleep(1)
        if n == 1:
            for j in range(10):
                print("\nChoices --> \n1. Snacks \n2. Drinks \n3. Desserts \n4. Check_out")
                sleep(0.5)
                ch = int(input("Choose any option from above --> "))
                sleep(1)
                if ch==1:
                    for k in range(10):
                        print("Snacks : ")
                        pprint(menu_card["Snacks"])
                        sleep(0.5)
                        cs = int(input("Choose any option from above --> "))
                        sleep(0.5)
                        if cs == 1:
                            s1 = int(input("How many Pizza to order --> "))
                            ps1 = s1*menu_card["Snacks"][ch][1]
                            orders.append(("Pizza",s1,ps1))
                            continue
                        elif  cs==2:
                            s2 = int(input("How many Burger to order --> "))
                            ps2 = s2*menu_card["Snacks"][ch][1]
                            orders.append(("Burger",s2,ps2))
                            continue
                        elif cs==3:
                            s3 = int(input("How many Veg_roll to order --> "))
                            ps3 = s3*menu_card["Snacks"][ch][1]
                            orders.append(("Veg_roll",s3,ps3))
                            continue
                        elif cs==4:
                            s4 = int(input("How many plats of Pasta to order --> "))
                            ps4 = s4*menu_card["Snacks"][ch][1]
                            orders.append(("Pasta",s4,ps4))
                            continue
                        elif cs==5:
                            s5 = int(input("How many plats of Chow_mein to order --> "))
                            ps5 = s5*menu_card["Snacks"][ch][1]
                            orders.append(("Chow_mein",s5,ps5))
                            continue
                        else:
                            sleep(1)
                            break
                elif ch==2:
                    for k in range(10):
                        print("Drinks : ")
                        pprint(menu_card["Drinks"])
                        sleep(0.5)
                        cd = int(input("Choose any option from above --> "))
                        sleep(0.5)
                        if cd == 1:
                            d1 = int(input("How many packets of Cold coffee to order --> "))
                            pd1 = d1*menu_card["Drinks"][cd][1]
                            orders.append(("Cold_coffee",d1,pd1))
                            continue
                        elif  cd==2:
                            d2 = int(input("How many cups of Hot coffee to order --> "))
                            pd2 = d2*menu_card["Drinks"][cd][1]
                            orders.append(("Hot_coffee",d2,pd2))
                            continue
                        elif cd==3:
                            d3 = int(input("How many botals of Coka cola to order --> "))
                            pd3 = d3*menu_card["Drinks"][cd][1]
                            orders.append(("Coka_cola",d3,pd3))
                            continue
                        else:
                            sleep(1)
                            break
                elif ch==3:
                    for k in range(10):
                        print("Desserts : ")
                        pprint(menu_card["Desserts"])
                        sleep(0.5)
                        ds = int(input("Choose any option from above --> "))
                        sleep(0.5)
                        if ds == 1:
                            ds1 = int(input("How many plats of Balu shahi to order --> "))
                            pds1 = ds1*menu_card["Desserts"][ds][1]
                            orders.append(("Balu_shahi",ds1,pds1))
                            continue
                        elif  ds==2:
                            ds2 = int(input("How much Gajar ka Halwa to order --> "))
                            pds2 = ds2*menu_card["Desserts"][ds][1]
                            orders.append(("Gajar_ka_Halwa",ds2,pds2))
                            continue
                        elif ds==3:
                            ds3 = int(input("How much Gulab jamun to order --> "))
                            pds3 = ds3*menu_card["Desserts"][ds][1]
                            orders.append(("Gulab_jamun",ds3,pds3))
                            continue
                        else:
                            sleep(1)
                            break
                elif ch==4:
                    if orders:
                        price = 0
                        food = []
                        time = "45 minutes"
                        for i in orders:
                            food.append((i[0],i[1]))
                            price += i[2]
                        
                        Id = 101
                        date = datetime.now().strftime("%x")
                        while date:
                            if date in Customers_record:
                                while Id:
                                    if Id in Customers_record[date]:
                                        Id+=1
                                    else:
                                        Customers_record[date][Id]={}
                                        Customers_record[date][Id]["Food"] = food
                                        Customers_record[date][Id]["Total_Price"] = price
                                        Customers_record[date][Id]["Delivery_Time"] = time
                                        break
                                break
                            else:
                                Customers_record[date] = {}
                                date = datetime.now().strftime("%x")
                        
                        print("-"*40)
                        print(f"\nOrderd food --> {food} \nTotal Price --> {price} \nFood Delivery Time --> {time}")
                        print(f"\n{'-'*40}")
                        sleep(0.5)
                    print("Exiting...")
                    sleep(1)
                    break
                else:
                    print("Choose right option...")
                    continue
            continue
                        
        elif n == 2:
            print("-"*40)
            print("\nMenu Card of Family Restaurant")
            pprint(menu_card)
            print(f"\n{'-'*40}")
            continue
            
        elif n == 3:
            date = datetime.now().strftime("%x")
            id_ = int(input("Enter your ID No. --> "))
            for i in range(5):
                if id_ in Customers_record[date]:
                    print("-"*40)
                    print("\nMy Cart")
                    print("\nOrderd Food -->",Customers_record[date][id_]["Food"])
                    print("Total Price --> ",Customers_record[date][id_]["Total_Price"])
                    print(f"\n{'-'*40}")
                    break
                else:
                    print("Please enter correct ID No.")
                    continue
            continue
            
        elif n == 4:
            print("Exiting...")
            sleep(2)
            break
        else:
            print("Please choose correct option...")
            continue
    return Customers_record
```
